```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from tabulate import tabulate
```

In this final evaluation , it is necessary aggregate the time of the 2 previous step Detector and descriptor

```python
## Importing report file.
df_knn_k1 = pd.read_csv('knn_output_k1/Matching_Descriptors_report.txt', sep=',',names=["Detector for k=1","Descriptor","image","Match","time","removed_keypoints"])

#Grouping values and converting columns from float to int
df_knn_k1_gr = df_knn_k1.groupby(["Detector for k=1"])["Match","removed_keypoints"].mean()
df_knn_k1_gr["removed_keypoints"] = df_knn_k1_gr["removed_keypoints"].astype(int)
df_knn_k1_gr["Match"] = df_knn_k1_gr["Match"].astype(int)
#print(tabulate(df_knn_k1_gr, tablefmt="github", headers="keys",numalign="center"))
df_knn_k1
```

```python
## Importing report file.
df_knn_k2 = pd.read_csv('knn_output_k2/Matching_Descriptors_report.txt', sep=',',names=["Detector for k=2","Descriptor","image","Match","time","removed_keypoints"])

#Grouping values and converting columns from float to int
df_knn_k2_gr = df_knn_k2.groupby(["Detector for k=2"])["Match","removed_keypoints"].mean()
df_knn_k2_gr["removed_keypoints"] = df_knn_k2_gr["removed_keypoints"].astype(int)
df_knn_k2_gr["Match"] = df_knn_k2_gr["Match"].astype(int)
#print(tabulate(df_knn_k2_gr, tablefmt="github", headers="keys",numalign="center"))
df_knn_k2
```
